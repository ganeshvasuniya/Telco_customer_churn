```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Telco_customer_churn.csv')

# Replace null values in 'Churn Reason' with a descriptive string
df['Churn Reason'] = df['Churn Reason'].fillna('Not Mentioned')

# Save the updated dataset
df.to_csv('Telco_Churn_Fixed_Reason.csv', index=False)
```

```python
# Check for fully identical rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Drop them if any exist (keeps the first occurrence, deletes the rest)
if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates removed successfully!")
else:
    print("No identical duplicates found. Moving forward!")

# Double check unique business keys (like CustomerID)
# Even if rows aren't 100% identical, two rows shouldn't have the same CustomerID
id_duplicates = df['CustomerID'].duplicated().sum()
print(f"Number of duplicate Customer IDs: {id_duplicates}")

if id_duplicates > 0:
    df = df.drop_duplicates(subset=['CustomerID'], keep='first')
```

```python
# Convert Total Charges to numeric
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors="coerce")
df["Total Charges"] = df["Total Charges"].fillna(0)
```

Feature Engineering

```python
# #Average Monthly Spending
df["AvgMonthlySpend"] = df["Total Charges"] / (df["Tenure Months"] + 1)
df["AvgMonthlySpend"] = df["AvgMonthlySpend"].round(2)
```

```python
# High Monthly Charge
median_charge = df["Monthly Charges"].median()
df["HighMonthlyCharge"] = (
    df["Monthly Charges"] > median_charge
).astype(int)
df[["Monthly Charges", "HighMonthlyCharge"]].head()
```

```python
# Long-Term Customer
df["LongTermCustomer"] = (
    df["Tenure Months"] >= 24
).astype(int)
df[["Tenure Months", "LongTermCustomer"]].head()
```

```python
# Auto Payment
df["AutoPayment"] = df["Payment Method"].str.contains(
    "automatic",
    case=False
).astype(int)
df[["Payment Method", "AutoPayment"]].head()
```
